```python
import pandas as pd
```

```python
dd=pd.read_csv("../datasets/daraz_dataset.csv")
# print(dd.shape)
print(dd.columns)
# print(dd.info())
# print(dd.describe())
print("max amount = ",dd["TotalAmount"].max())
print('minimun amount =',dd["TotalAmount"].min())
print("avarage amount = ",dd["TotalAmount"].mean())
print(dd[["CustomerName","Age","TotalAmount"]])
```

iloc = integer location
we can acess dataset with integer number 

```python
#print(dd.iloc[0])
print(dd.iloc[11]) #11 index row 
```

```python
print(dd.iloc[0:5])
```

```python
#iloc[row,colum] to get specific 
#iloc[row:row] to get row to row
print(dd.iloc[0,2])  
```

```python
#print(dd.iloc[:]) #to print all row and column
print(dd.iloc[:,2])  # to get 3 column from all row
```

```python
print(dd.iloc[:,[2,3,4]])  #its get 2,3,4 column from all row
```

```python
print(dd.iloc[-1])
#to get last row
```

```python
print(dd.iloc[0:3,0:3]) # its print 3 row and 3 column 
```

```python
print(dd.loc[:,["CustomerName","Age"]])  #read with loc dd.loc[row:,column name]
```

```python
print(dd[dd["District"]=="Kathmandu"])  #get customer info if district = Kathamandu
```

```python
print(dd.loc[dd['TotalAmount']>=30000])
```

```python
import pandas as pd
dd = pd.read_csv("../datasets/daraz_dataset.csv")

```

```python
#1
print(dd[dd["District"]=="Kathmandu"])
```

```python
#2
print(dd[dd["TotalAmount"]>50000])
```

```python
#3
print(dd[dd["Age"]>30])
```

```python
#4
print(dd[["CustomerName","TotalAmount"]])
```

```python
#5
print(dd.iloc[:10])
```

```python
#6
print(dd[(dd["District"]=="Kathmandu") &(dd["TotalAmount"]>=30000)])
```

```python
#7
print(dd[dd["District"].isin(["Kathmandu","Pokhara"])])
```

```python
#8
print(dd[dd["Age"].between(20,30)])
```

```python
#9
print(dd[(dd["PaymentMethod"]=="COD")&(dd["TotalAmount"]>20000)])
```

```python
#10
print(dd[dd["District"]!="Kathmandu"])
```

```python
#11
print(dd.loc[dd["TotalAmount"]>30000,["CustomerName","District"]])
```

```python
#12
print(dd[dd["District"].isin(["Kathmandu","Lalitpur","Bhaktapur"])])
```

```python
#13
print(dd.sort_values("TotalAmount",ascending=False).head(5))
```

```python
#14
print(dd[(dd["District"]=="Kathmandu")&(dd["Age"]>25)&(dd["TotalAmount"]>50000)])
```

```python
#15
print(dd.loc[dd["TotalAmount"].between(10000,50000)])
```

```python
#16
print(dd.loc[dd["PaymentMethod"].isin(["COD","eSewa"])])
```

```python
#17
print(dd.sort_values("TotalAmount",ascending=False).head(5))
```

```python
#18
print(dd[dd["CustomerName"].str.startswith("A")])
```

```python
#19
print(dd[dd["CustomerName"].str.endswith("a")])
```

```python
#20
print(dd[dd["CustomerName"].str.contains("an")])
```

```python
#21
print(dd["District"].value_counts())
```

```python
print(dd.groupby("District")["TotalAmount"].mean())
```

```python
#23
print(dd["TotalAmount"].max())
print(dd.loc[dd["TotalAmount"].idxmax(),["CustomerName","Age"]])
#idx to find the index of that
```
